```python
import random
import time


# --------------------- Data Input ---------------------

def manual_input():
    n = int(input("Enter number of elements: "))
    arr = []

    print("Enter elements in sorted order:")
    for i in range(n):
        arr.append(int(input(f"Element {i + 1}: ")))

    return arr


def random_uniform():
    n = int(input("Enter number of elements: "))
    start = int(input("Enter starting value: "))
    step = int(input("Enter uniform gap (step): "))

    arr = [start + i * step for i in range(n)]
    return arr


def random_non_uniform():
    n = int(input("Enter number of elements: "))

    arr = []
    value = random.randint(1, 10)

    for _ in range(n):
        value += random.randint(1, 20)   # Random gap
        arr.append(value)

    return arr


# --------------------- Interpolation Search ---------------------

def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    start_time = time.perf_counter()

    while low <= high and key >= arr[low] and key <= arr[high]:

        probes += 1

        if arr[low] == arr[high]:
            if arr[low] == key:
                end_time = time.perf_counter()
                return low, probes, end_time - start_time
            break

        pos = low + int(
            ((key - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if pos < low or pos > high:
            break

        if arr[pos] == key:
            end_time = time.perf_counter()
            return pos, probes, end_time - start_time

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    end_time = time.perf_counter()
    return -1, probes, end_time - start_time


# --------------------- Dataset Selection ---------------------

def create_dataset(choice):
    print("\n1. Enter Manually")
    print("2. Generate Randomly")

    option = int(input("Choose option: "))

    if choice == 1:
        if option == 1:
            arr = manual_input()
        else:
            arr = random_uniform()

    else:
        if option == 1:
            arr = manual_input()
        else:
            arr = random_non_uniform()

    arr.sort()
    return arr


# --------------------- Main Program ---------------------

def main():

    print("Interpolation Search on User-Defined Dataset")
    print("--------------------------------------------")
    print("1. Uniform Dataset")
    print("2. Non-Uniform Dataset")

    dataset_choice = int(input("Choose dataset type: "))

    dataset = create_dataset(dataset_choice)

    print("\nDataset:")
    print(dataset)

    key = int(input("\nEnter search key: "))

    index, probes, exec_time = interpolation_search(dataset, key)

    print("\n========== Result ==========")

    if index != -1:
        print("Key Found")
        print("Index Position :", index)
    else:
        print("Key Not Found")

    print("Number of Probes :", probes)
    print(f"Execution Time : {exec_time:.10f} seconds")

    print("\nPerformance Observation")

    if dataset_choice == 1:
        print("- Uniform datasets usually require fewer probes.")
        print("- Average Time Complexity: O(log log n)")
    else:
        print("- Non-uniform datasets may require more probes.")
        print("- Worst-case Time Complexity can approach O(n)")


if __name__ == "__main__":
    main()
```
